```python
import scrapy
import pandas as pd
from scrapy.crawler import CrawlerProcess
```

```python
class CompanySpider(scrapy.Spider):
    name = 'company'
    
    def start_requests(self):
        urls = [
            'https://jncelektrotechniek.nl/',
            'https://stwarmte.nl/',
            # add more URLs here
        ]
        for url in urls:
            yield scrapy.Request(url=url, callback=self.parse)

    def parse(self, response):
        item = {
            'name': response.css('div.company-name::text').get(),
            'address': response.css('div.company-address::text').get(),
            'phone': response.css('div.company-phone::text').get(),
            # add more fields as necessary
        }
        self.items.append(item)
        yield item
        
    def __init__(self, *args, **kwargs):
        super(CompanySpider, self).__init__(*args, **kwargs)
        self.items = []

    def closed(self, reason):
        df = pd.DataFrame(self.items)
        df.to_excel('companiedata.xlsx', index=False)    
```

```python
# settings.py

BOT_NAME = 'company_info'

SPIDER_MODULES = ['company_info.spiders']
NEWSPIDER_MODULE = 'company_info.spiders'

# Obey robots.txt rules (can be set to False if necessary)
ROBOTSTXT_OBEY = True

# Configure item pipelines (if using)
ITEM_PIPELINES = {
   'company_info.pipelines.CompanyInfoPipeline': 300,
}
```

```python
process = CrawlerProcess()
process.crawl(CompanySpider)
process.start()
```
